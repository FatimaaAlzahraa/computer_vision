```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

preprocessing

```python
image = cv2.imread ("/content/Macaws.jpg")
```

```python
plt.imshow (image)
```

```python
increasing_image = cv2.resize (image , (1500, 1300))  #width , hight
descreasing_image = cv2.resize (image , (100, 200))
```

```python
plt.imshow (increasing_image)
```

```python
plt.imshow (descreasing_image)
```

```python
clockwise_rotate = cv2.rotate (image , cv2.ROTATE_90_CLOCKWISE)
plt.imshow (clockwise_rotate)
```

```python
#based noise in image
smoothen = cv2.blur (image , (50,70))
plt.imshow (smoothen)
```

```python
#line in image
start_point = (0,0)
end_point = (500,500)
blue_color = (0,0,255)
line_thickness = (12)
line = cv2.line(image , start_point , end_point , blue_color ,line_thickness )
plt.imshow (line)
```

```python
#convert image grayScale
gray = cv2.cvtColor(image , cv2.COLOR_BGRA2GRAY)
plt.imshow (gray)
```

```python
#detection edges in image
edges1 = cv2.Canny (image , 300 , 300)  #dark  # idintify max and min threshold
plt.imshow (edges1)
```

```python
edges2 = cv2.Canny (image , 75 , 75)  #light
plt.imshow (edges2)
```

```python
ret, binary_thresholding = cv2.threshold (image , 120,255,cv2.THRESH_BINARY)
plt.imshow (binary_thresholding)
```

```python
image1 = cv2.imread ("/content/Macaws.jpg")
gray_scale = cv2.cvtColor(image1 , cv2.COLOR_BGRA2GRAY)
edges = cv2.Canny (image1 , 30 , 200)  #image  #diminsion
```

```python
coutours, hieriarchy = cv2.findContours (edges, cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
```

```python
cv2.drawContours (image1, coutours, -1 , (255,0,0), 3)
```

```python
plt.imshow (image1)
```

```python
kernal = np.ones ((5,5), np.uint8 )
```

erosion and dilation to decrease noise

```python
image_erosion = cv2.erode (image, kernal, iterations=1 ) #dark
plt.imshow (image_erosion)
```

```python
image_dilation = cv2.dilate (image, kernal, iterations=1 ) #light
plt.imshow (image_dilation)
```
